```python
import numpy as np 
import pandas as pd
import os
```

```python
nw1 = pd.read_csv("R_Module_Day_12.1_Network_Intrusion_Train_data.csv")
```

```python
nw2 = pd.read_csv("R_Module_Day_12.3_Network_Intrusion_Validate_data.csv")
```

```python
nw2.isnull().sum()
```

```python
# Do imputations

nw1['protocol_type'] = nw1['protocol_type'].astype('category')
nw1['protocol_type'] = nw1['protocol_type'].cat.codes
nw1['service'] = nw1['service'].astype('category')
nw1['service'] = nw1['service'].cat.codes
nw1['flag'] = nw1['flag'].astype('category')
nw1['flag'] = nw1['flag'].cat.codes
```

```python
nw2['protocol_type'] = nw2['protocol_type'].astype('category')
nw2['protocol_type'] = nw2['protocol_type'].cat.codes
nw2['service'] = nw2['service'].astype('category')
nw2['service'] = nw2['service'].cat.codes
nw2['flag'] = nw2['flag'].astype('category')
nw2['flag'] = nw2['flag'].cat.codes
```

```python
nw1.dtypes
```

```python
len(nw1.columns)
```

```python
# Split into i/p and 0/p

train_x = nw1.iloc[:,0:41]
train_y = nw1.iloc[:,41:]
```

```python
train_y.head()
```

```python
val_x = nw2.iloc[:,0:41]
val_y = nw2.iloc[:,41:]
```

```python
train_x.head()
```

```python
val_x.groupby('flag').agg({'flag':np.size})
```

```python
# Make dummies  in train dataset

train_x = pd.get_dummies(train_x,columns = ['protocol_type'],drop_first = True)
```

```python
train_x.head()
```

```python
train_x = pd.get_dummies(train_x,columns = ['flag'],drop_first = True)
```

```python
train_x.head()
```

```python
# Make tarin dataset into array

x_train_x = train_x.iloc[:,:].values
```

```python
# Make dummies in test dataset

val_x = pd.get_dummies(val_x,columns = ['protocol_type'],drop_first = True)
```

```python
val_x = pd.get_dummies(val_x,columns = ['flag'],drop_first = True)
```

```python
val_x.head()
```

```python
x_val_x = val_x.iloc[:,:].values
```

```python
x_val_x
```

```python
train_y['class'] = train_y['class'].astype('category')
train_y['class'] = train_y['class'].cat.codes
```

```python
val_y['class'] = val_y['class'].astype('category')
val_y['class'] = val_y['class'].cat.codes
```

```python
train_y.head()
```

```python
y_train_y = train_y.iloc[:,:].values
```

```python
y_val_y = val_y.iloc[:,:].values
```

```python
y_train_y
```

```python
# Now do scaling

from sklearn.preprocessing import StandardScaler
stdsc = StandardScaler()
x_train_x= stdsc.fit_transform(x_train_x)
x_val_x = stdsc.transform(x_val_x)
```

```python
# Making decision tree model

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy',random_state = 0)
dtc.fit(x_train_x,y_train_y)
```

```python
# Do prediction 

y_pred_DT = dtc.predict (x_val_x)
```

```python
# find confusion matrix and accurecy of the model

from sklearn.metrics import confusion_matrix,accuracy_score
cm_DT = confusion_matrix(y_pred_DT,y_val_y)
acc_DT = accuracy_score(y_pred_DT,y_val_y)
print("confusion matrix is\n %s"%cm_DT)
print("accuracy of decision tree model is %s"%acc_DT)
```

```python
# Make random forest model

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 50,random_state = 0, criterion = 'entropy')
rfc.fit(x_train_x,y_train_y)
```

```python
y_pred_rfc = rfc.predict(x_val_x)
```

```python
cm_rfc =confusion_matrix(y_val_y,y_pred_rfc)
acc_rfc = accuracy_score(y_val_y,y_pred_rfc)
print("confusion matrix is\n %s"%cm_rfc)
print("accuracy of decision tree model is %s"%acc_rfc)
```
